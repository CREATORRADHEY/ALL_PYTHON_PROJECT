```
attempts = 0 # Initialize attempts counter
MAX_ATTEMPTS = 3
CORRECT_PIN = 4567
INITIAL_BALANCE = 50000
current_balance = INITIAL_BALANCE

# Loop continues as long as attempts are less than the maximum limit
while attempts < MAX_ATTEMPTS :
  user_pin = int (input ("Enter your pin :  ") )

  if user_pin == CORRECT_PIN :
    print ("\nLogin Successful! ðŸ§")
    print ("Your current balance is : " , current_balance)

    # Transaction logic...
    amount = int (input ("Enter the amount you want to withdraw : "))
    if amount <= current_balance:
      current_balance -= amount
      print (f"\nCongrats! Amount of {amount} withdrawn. ðŸŽ‰")
      print ("New balance is : " , current_balance )
    else:
      print("\nError: Insufficient balance.")

    break # Exit the loop after a successful transaction

  else :
    attempts += 1 # Increment attempt counter

    # Check if attempts remain
    if attempts < MAX_ATTEMPTS :
      remaining = MAX_ATTEMPTS - attempts
      print (f"\nIncorrect PIN. You have {remaining} attempts remaining.")
    else:
      # This block executes when the last attempt is also incorrect
      print (f"\n**Access Blocked.** ðŸ”’ You have exceeded the maximum of {MAX_ATTEMPTS} incorrect attempts.")
      break # Exit the loop because access is blocked

print("\nThank you for using the ATM!")
```
